investors assuming credit risk.

## Feature Engineering

```python
# Replace Missing Values with Median for Numerical Columns and Mode for Categorical Columns
numerical_cols = Finance_Data.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = Finance_Data.select_dtypes(include=['object']).columns.tolist()  # Corrected this line

for col in numerical_cols:
    Finance_Data[col].fillna(Finance_Data[col].median(), inplace=True)
```

```python
for col in categorical_cols:
    Finance_Data[col].fillna(Finance_Data[col].mode()[0], inplace=True)
```

***fillna()***: The fillna() method in pandas is used to fill missing (NaN) values in a DataFrame or Series. Here, Finance_Data[col].mode()[0] calculates the mode (most frequent value) of the current categorical column, and this mode value is used to fill missing values.

***inplace=True***: The inplace=True parameter modifies the original DataFrame (Finance_Data) in place, without creating a new DataFrame.


```python
label_encoder = LabelEncoder()

for col in categorical_cols:
    Finance_Data[col] = label_encoder.fit_transform(Finance_Data[col])
```

***LabelEncoder()***: The LabelEncoder class in scikit-learn is used to encode categorical string or integer labels into numerical labels (0, 1, 2, ...).

***categorical_cols***: This loop iterates over each column specified in the categorical_cols list, which contains the names of categorical columns in the Finance_Data DataFrame.

***fit_transform()***: The fit_transform() method in scikit-learn fits the LabelEncoder on the column's unique labels and transforms the categorical values into numerical labels.

The ***label_binarize()*** function in scikit-learn is used to convert a multi-class target variable into a binary format. By applying this code, you:

```python
# Compute ROC Curve & ROC Area for Each Class
from sklearn.metrics import roc_curve, auc
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

The ***roc_curve()*** function in scikit-learn computes the Receiver Operating Characteristic (ROC) curve by plotting the true positive rate (Sensitivity) against the false positive rate (1-Specificity) at various threshold settings. It requires the true binary labels and the predicted probabilities for a specific class to calculate the ROC curve.

The ***auc()*** function in scikit-learn calculates the Area Under the Curve (AUC) of the ROC curve, which quantifies the overall performance of the classifier in distinguishing between the positive and negative classes across various threshold values. A higher AUC value indicates better classifier performance.

```python
# Plot ROC Curve for Each Class
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
colors = ['blue', 'red']
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

## Loan Approving Strategy

The loan approving strategy is crucial in determining which loan applications should be approved based on the risk category and credit score. In the provided code, the loan approving strategy is already implemented as a function named loan_approval_strategy.

Here's a brief overview of the loan approving strategy:

**Loan Approving Strategy:**

**Low Risk:** All loan applications categorized as 'Low Risk' are automatically approved.

**Medium Risk:** Loan applications categorized as 'Medium Risk' are approved only if the credit score is 70 or above.

**High Risk:** All loan applications categorized as 'High Risk' are declined.

```python
credit_score_data = pd.read_csv('Credit_Score.csv')
F_D = pd.merge(Finance_Data, credit_score_data, on='member_id')
```

```python
# Assuming You Have a Dictionary of Credit Scores for Each member_id
credit_scores = {'member_id': [1, 2, 3], 'Credit_Score': [700, 650, 720]}
```

```python
# Create a New DataFrame for the Credit Scores
credit_scores_df = pd.DataFrame(credit_scores)
```

```python
# Merge the Credit_Scores DataFrame with the Finance_Data DataFrame
F_D = F_D.merge(credit_scores_df, on='member_id')
```

```python
def loan_approval_strategy(row):
    if row['dti'] < 0.3 and row['delinq_2yrs'] == 0 and row['avg_credit_score'] >= 70:
        return 'Approved'
    elif row['addr_state'] in ['Medium Risk', 'Low Risk'] and row['avg_credit_score'] >= 65:
        return 'Approved'
    else:
        return 'Declined'
```

```python
F_D['loan_approval_status'] = F_D.apply(loan_approval_strategy, axis=1)
F_D.to_csv('F_D.csv', index=False)
```

## Components of Loan Approving Strategy:

**Risk Categorization:**

Low Risk: Borrowers with a strong credit history, stable income, and low debt-to-income ratio.

Medium Risk: Borrowers with some credit issues or moderate debt-to-income ratio.

High Risk: Borrowers with poor credit history, high debt-to-income ratio, or other financial red flags.

**Credit Scoring:** Assessing the creditworthiness of borrowers based on their credit score, which is calculated using various factors such as payment history, credit utilization, and length of credit history.

**Loan-to-Income Ratio:** Evaluating the borrower's ability to repay the loan by comparing the loan amount to their income.

**Debt-to-Income Ratio:** Analyzing the borrower's financial stability by comparing their total debt to their income.
